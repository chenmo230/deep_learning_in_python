```python
import numpy as np
```

## 1、implement sigmoid function



```python
def basic_sigmoid(x):
    """
    计算sigmoid函数
    """
    return 1 / (1 + np.exp(-x))
```

```python
tensor_1 = np.array([-20, -10, -1, 0, 1, 10, 20])
basic_sigmoid(tensor_1)
```

```python
import matplotlib.pyplot as plt

# draw basic_sigmoid gragh
start = -10
stop = 10
step = 0.01
num = (stop - start) / step
x = np.linspace(start, stop, int(num))
y = basic_sigmoid(x)

plt.plot(x, y, label='basic_sigmoid')
plt.grid(True)
plt.legend()
plt.show()
```

## 2、implement sigmoid gradient

$\sigma'(x) =\sigma(x)(1-\sigma(x))$

```python
def sigmoid_derivative(x):
    s = basic_sigmoid(x)
    ds = s * (1 - s)
    
    return ds
```

```python
sigmoid_derivative(tensor_1)
```

## 3、implement tanh function




```python
def basic_tanh(x):
    return (1 - np.exp(-2*x)) / (1 + np.exp(-2*x))
```

```python
basic_tanh(tensor_1)
```

```python
# draw basic_tanh gragh

start = -10
stop = 10
step = 0.01
num = (stop - start) / step
x = np.linspace(start, stop, int(num))
y = basic_tanh(x)

plt.plot(x, y, label='basic_tanh')
plt.grid(True)
plt.legend()
plt.show()
```

## 4、tanh gradient


```python
def tanh_derivative(x):
    tanh = basic_tanh(x)
    ds = 1 - tanh ** 2
    
    return ds
```

```python
tanh_derivative(tensor_1)
```

## 5、implement relu



```python
def basic_relu(x):
    return np.maximum(0, x)
```

```python
basic_relu(tensor_1)
```

```python
# draw basic_relu gragh
start = -10
stop = 10
step = 0.01
num = (stop - start) / step
x = np.linspace(start, stop, int(num))
y = basic_relu(x)

plt.plot(x, y, label='basic_relu')
plt.grid(True)
plt.legend()
plt.show()
```

## 6、implement relu gradient


```python
def relu_derivative(x):
    return np.where(x > 0, 1, 0)
```

```python
relu_derivative(tensor_1)
```

## 7、implement Leaky relu



```python
def basic_leaky_relu(x):
    return np.maximum(0.01*x, x)
```

```python
basic_leaky_relu(tensor_1)
```

```python
# draw basic_leaky_relu gragh
start = -10
stop = 10
step = 0.01
num = (stop - start) / step
x = np.linspace(start, stop, int(num))
y = basic_leaky_relu(x)

plt.plot(x, y, label='basic_leaky_relu')
plt.grid(True)
plt.legend()
plt.show()
```

## 8、implement leaky relu gradient


```python
def leaky_relu_derivative(x):
    return np.where(x > 0, 1, 0.01)
```

```python
leaky_relu_derivative(tensor_1)
```
